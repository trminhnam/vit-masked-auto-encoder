```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

FOLDER_PATH_INP = "/content/drive/MyDrive/VDT - Nhóm 6 - Module 2 /ViT-MaskedAttention/data/Data_ver1/Data"
FOLDER_PATH_OUT = "/content/OUT"
categories =  ["Awake", "Tired"]
```

```python
def crop_face(img, out_path):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30), flags=cv2.CASCADE_SCALE_IMAGE)

    if(len(faces) == 0):
        return
    x, y, w, h = faces[0]
    # calculate the center of the face
    center_x = x + w // 2
    center_y = y + h // 2
    
    # calculate the size of the rectangle
    rect_size = 72
    
    # calculate the top-left corner of the rectangle
    rect_x = center_x - rect_size // 2
    rect_y = center_y - rect_size // 2
    
    # make sure the rectangle is within the bounds of the image
    rect_x = max(0, rect_x)
    rect_y = max(0, rect_y)
    rect_x2 = min(img.shape[1], rect_x + rect_size)
    rect_y2 = min(img.shape[0], rect_y + rect_size)
                
    rect_img = img[rect_y:rect_y2, rect_x:rect_x2]

    cv2.imwrite(out_name, rect_img)
```

```python
categories = ["Tired"]
```

```python
for categorie in categories:
    for folder in os.listdir(f'{FOLDER_PATH_INP}/{categorie}'):
        for img_path in os.listdir(f'{FOLDER_PATH_INP}/{categorie}/{folder}'):

            in_path = f'{FOLDER_PATH_INP}/{categorie}/{folder}/{img_path}'
            out_name = f'{FOLDER_PATH_OUT}/{categorie}/{folder}_{img_path}'

            print(out_name)

            img = cv2.imread(in_path)
            # plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
            # plt.show()
            crop_face(img, out_name)
```
