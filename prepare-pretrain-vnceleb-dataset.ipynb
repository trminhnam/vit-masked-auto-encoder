```python
import os
import cv2
import numpy
from tqdm.auto import tqdm
import matplotlib.pyplot as plt

from utils.face_alignment.align_faces import face_detection, align_face
```

```python
vn_celeb_dataset_dir = '/content/drive/MyDrive/ViT-MaskedAttention/data/VN-celeb'
pretrain_img_dir = '/content/drive/MyDrive/ViT-MaskedAttention/data/pretrain'
```

```python
for img_fold_num in tqdm(os.listdir(vn_celeb_dataset_dir)):
    img_fold_dir = os.path.join(vn_celeb_dataset_dir, img_fold_num)
    if not os.path.isdir(img_fold_dir):
        if img_fold_dir.endswith('.png'):
            os.remove(img_fold_dir)
            print(f'remove {img_fold_dir}')
        continue
    for img_name in os.listdir(img_fold_dir):
        img_path = os.path.join(img_fold_dir, img_name)
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        img_save_name = f'{img_fold_num}_{img_name}'
        img_save_path = os.path.join(pretrain_img_dir, img_save_name)
        if os.path.exists(img_save_path):
            continue
            
        try:
            alignedFace = align_face(img)
            new_img, gray_img = face_detection(alignedFace)
            # plt.imshow(new_img)
            # plt.show()
            # print(f'{result} for img {img_save_name}')
        except Exception as e:
            print(e)

        new_img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        result = cv2.imwrite(img_save_path, new_img)
```
